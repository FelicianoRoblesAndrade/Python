```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import zipfile
import urllib.request
import zipfile
url = 'http://www.dgis.salud.gob.mx/descargas/datosabiertos/recursosSalud/Recursos_Secretaria_Salud_2023.zip?V=2024.04.30'

filename = 'filename.zip'

urllib.request.urlretrieve(url, filename)
with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall('.')
```

```python
# Consultamos el csv de interes
# descriptores_url = "http://www.dgis.salud.gob.mx/descargas/datosabiertos/recursosSalud/Descriptores_Recursos.xlsx?V=1.1"
# df_des = pd.read_excel(descriptores_url)
# De este excel pudimos discriminar los datos clave de interes (aquellos que inicien dentro del rango "E13-E25" y "C1706-C1750")
```

```python
local="/content/Recursos Secretaria de Salud 2023.csv"
df=pd.read_csv(local, engine="python",encoding='latin-1')
df.info() # Vemos la informacion del dataframe
```

```python
df1=df[['ENTIDAD',"E13",'E18','E24']]
df1=df1.rename(columns={'ENTIDAD':'entidad', 'E13':'consultorios','E18':'medicos_GrEsOd','E24':'personal_tecnico'})
df1.head()
```

```python
df2=df1.groupby(['entidad'])[['consultorios','medicos_GrEsOd','personal_tecnico']].mean()
df2=df2.reset_index()
df2.head(10)
```

```python
print(f"Hay {len(df2)} entidades diferentes")
```

```python
for i in df2:
  print(i)
```

```python
for i in range(len(df2)):
  for i2 in df2:
    if df2[i2][i]==df2[i2].max():
        if i2=="entidad":
          continue
        else:
          print(f'Estado con mayor(es) {i2}:\n{df2["entidad"][i]}\n')
    if df2[i2][i]==df2[i2].min():
        if i2=="entidad":
          continue
        else:
          print(f'Estado con menor(es) {i2}:\n{df2["entidad"][i]}\n')
```
